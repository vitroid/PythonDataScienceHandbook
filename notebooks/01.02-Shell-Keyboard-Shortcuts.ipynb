# IPython シェルのキーボード ショートカット

コンピューターである程度の時間を過ごす場合は、ワークフローでキーボード ショートカットを使用している可能性があります。
さまざまなプログラムやシステムでコピー アンド ペーストするための Cmd-C と Cmd-V (または Ctrl-C と Ctrl-V) は、おそらく最もよく知られています。
パワー ユーザーはさらに先へ進む傾向があります。Emacs、Vim などの一般的なテキスト エディターは、複雑なキーストロークの組み合わせによって、信じられないほど幅広い操作をユーザーに提供します。

IPython シェルはここまでは行きませんが、コマンドを入力する際に​​すばやく移動できるように、多数のキーボード ショートカットを提供しています。
これらのショートカットは、実際には IPython 自体によって提供されるのではなく、GNU Readline ライブラリへの依存を通じて提供されます。そのため、次のショートカットの一部は、システム構成によって異なる場合があります。
また、これらのショートカットの一部はブラウザー ベースのノートブックで機能しますが、このセクションでは主に IPython シェルのショートカットについて説明します。

これらに慣れると、「ホーム」キーボードの位置から手を動かさずに特定のコマンドをすばやく実行するのに非常に役立ちます。
Emacs のユーザーであるか、Linux スタイルのシェルの経験がある場合は、次のことは非常によく知られています。
これらのショートカットをいくつかのカテゴリにグループ化します: *ナビゲーション ショートカット*、*テキスト入力ショートカット*、*コマンド履歴ショートカット*、*その他のショートカット*。

## ナビゲーション ショートカット

行内を前後に移動するために左矢印キーと右矢印キーを使用することは明らかですが、「ホーム」キーボード位置から手を動かす必要のない他のオプションがあります。

| |キーストローク |アクション | アクション |
|-----------------------------------|--------------------------------------------|
| | Ctrl-a |カーソルを行頭に移動 |
| | ``Ctrl-e`` |カーソルを行末に移動 |
| | ``Ctrl-b`` または左矢印キー |カーソルを 1 文字戻す |
| | ``Ctrl-f`` または右矢印キー |カーソルを 1 文字進めます |

## テキスト入力のショートカット

Backspace キーを使用して前の文字を削除することは誰もが慣れていますが、キーに手を伸ばすには、多くの場合、ちょっとした指の体操が必要で、一度に 1 文字しか削除されません。
IPython には、入力中のテキストの一部を削除するためのショートカットがいくつかあります。
これらの中で最もすぐに役立つのは、テキスト行全体を削除するコマンドです。
Backspace キーを押して前の文字を削除するのではなく、Ctrl-b と Ctrl-d の組み合わせを使用していることに気付いた場合、これらが第二の性質になっていることがわかります。

| |キーストローク |アクション | アクション |
|-------------------------------|--------------------------------------------------|
| |バックスペース キー |行の前の文字を削除 |
| | ``Ctrl-d`` |行の次の文字を削除 |
| | Ctrl-k |カーソル位置から行末までテキストを切り取る |
| | Ctrl-u |行頭からカーソルまでのテキストを切り取る |
| | ``Ctrl-y`` |以前に切り取られたテキストをヤンク (貼り付け) |
| | Ctrl-t |前の 2 文字を入れ替える (つまり、入れ替える) |

## コマンド履歴のショートカット

おそらく、ここで説明する最も影響力のあるショートカットは、コマンド履歴をナビゲートするために IPython が提供するショートカットです。
このコマンド履歴は、現在の IPython セッションを超えるものです。コマンド履歴全体は、IPython プロファイル ディレクトリの SQLite データベースに保存されます。
これらにアクセスする最も簡単な方法は、上下の矢印キーを使用して履歴を順に表示することですが、他のオプションもあります。

| |キーストローク |アクション | アクション |
|-------------------------------------|--------------------------------------------|
| | ``Ctrl-p`` (または上矢印キー) |履歴の前のコマンドにアクセス |
| | ``Ctrl-n`` (または下矢印キー) |履歴の次のコマンドにアクセス |
| | ``Ctrl-r`` |コマンド履歴の逆検索 |

逆検索は特に便利です。
前のセクションで ``square`` という関数を定義したことを思い出してください。
新しい IPython シェルから Python の履歴を逆検索して、この定義をもう一度見つけてみましょう。
IPython ターミナルで Ctrl-r を押すと、次のプロンプトが表示されます。

```ipython
In [1]:
(reverse-i-search)`': 
```

このプロンプトで文字の入力を開始すると、IPython はそれらの文字に一致する最新のコマンドがあればそれを自動入力します。

```ipython
In [1]: 
(reverse-i-search)`sqa': square??
```

いつでも、さらに文字を追加して検索を絞り込むか、もう一度 Ctrl-r を押して、クエリに一致する別のコマンドをさらに検索できます。前のセクションに従っている場合、Ctrl-r をさらに 2 回押すと、次のようになります。

```ipython
In [1]: 
(reverse-i-search)`sqa': def square(a):
    """Return the square of a"""
    return a ** 2
```

探しているコマンドが見つかったら、Return キーを押すと検索が終了します。
次に、取得したコマンドを使用して、セッションを続行できます。

```ipython
In [1]: def square(a):
    """Return the square of a"""
    return a ** 2

In [2]: square(2)
Out[2]: 4
```

Ctrl-p/Ctrl-n または上/下矢印キーを使用して履歴を検索することもできますが、行頭の文字に一致する場合のみです。
つまり、**``def``** と入力してから Ctrl-p を押すと、履歴の中で文字 ``def`` で始まる最新のコマンド (存在する場合) が検索されます。

## その他のショートカット

最後に、上記のどのカテゴリにも当てはまらないが、知っておくと便利なその他のショートカットがいくつかあります。

| |キーストローク |アクション | アクション |
|-------------------------------|--------------------------------------------|
| | Ctrl-l |端末画面クリア |
| | ``Ctrl-c`` |現在の Python コマンドを中断 |
| | ``Ctrl-d`` | IPython セッションを終了 |

特に Ctrl-c は、非常に長時間実行されるジョブを誤って開始した場合に役立ちます。

ここで説明するショートカットのいくつかは、最初は少し面倒に思えるかもしれませんが、練習すればすぐに自動化されます。
その筋肉の記憶を発達させると、他の状況でそれらが利用可能になりたいと思うことさえあると思います.
