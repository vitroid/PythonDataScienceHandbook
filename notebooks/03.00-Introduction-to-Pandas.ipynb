# Pandas によるデータ操作

前の章では、NumPy とその ``ndarray`` オブジェクトについて詳しく説明しました。これは、Python で密な型付き配列の効率的な格納と操作を提供します。
ここでは、Pandas ライブラリによって提供されるデータ構造を詳細に調べることで、この知識に基づいて構築します。
Pandas は NumPy の上に構築された新しいパッケージであり、効率的な ``DataFrame`` の実装を提供します。
``DataFrame`` は基本的に、行と列のラベルが付加された多次元配列であり、多くの場合、異種の型や欠損データが含まれています。
Pandas は、ラベル付けされたデータ用の便利なストレージ インターフェイスを提供するだけでなく、データベース フレームワークとスプレッドシート プログラムの両方のユーザーになじみのある強力なデータ操作を多数実装しています。

これまで見てきたように、NumPy の ``ndarray`` データ構造は、数値計算タスクで通常見られる、クリーンでよく整理されたデータのタイプに不可欠な機能を提供します。
この目的には非常に役立ちますが、より柔軟性が必要な場合 (例: データへのラベルの添付、欠落データの操作など)、および要素単位のブロードキャストにうまく対応しない操作を試みる場合 (例:グループ化、ピボットなど)、これらはそれぞれ、私たちの周りの世界でさまざまな形式で利用可能な構造化されていないデータを分析する上で重要な要素です。
Pandas、特にその ``Series`` および ``DataFrame`` オブジェクトは、NumPy 配列構造に基づいて構築されており、データ サイエンティストの多くの時間を占めるこの種の「データ変更」タスクへの効率的なアクセスを提供します。

この章では、「Series」、「DataFrame」、および関連する構造を効果的に使用する仕組みに焦点を当てます。
必要に応じて実際のデータセットから引き出された例を使用しますが、これらの例は必ずしも焦点ではありません。

このインストールの詳細については、[Pandas documentation](http://pandas.pydata.org/) を参照してください。

````python
import pandas
pandas.__version__
````
