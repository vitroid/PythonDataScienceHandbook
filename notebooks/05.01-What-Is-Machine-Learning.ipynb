# 機械学習とは

さまざまな機械学習手法の詳細を見ていく前に、まず機械学習とは何か、そうでないものを見ていきましょう。
機械学習は人工知能のサブフィールドとして分類されることがよくありますが、一見しただけでは分類が誤解を招くことがよくあります。
機械学習の研究は、確かにこの文脈での研究から生まれましたが、機械学習手法のデータ サイエンスへの応用においては、機械学習を *データのモデルを構築する* 手段と考える方が役に立ちます。

基本的に、機械学習には、データの理解に役立つ数学的モデルの構築が含まれます。
これらのモデルに、観察されたデータに適応できる*調整可能なパラメーター*を与えると、「学習」が争いに加わります。このように、プログラムはデータから「学習」していると見なすことができます。
これらのモデルが以前に見られたデータに適合すると、それらを使用して、新しく観察されたデータの側面を予測および理解できます。
このタイプの数学的モデルベースの「学習」が、人間の脳によって示される「学習」にどの程度似ているかについて、より哲学的な余談を読者に残しておきます。

機械学習の問題設定を理解することは、これらのツールを効果的に使用するために不可欠です。そのため、ここで説明するアプローチの種類を大まかに分類することから始めます。

## 機械学習のカテゴリ

最も基本的なレベルでは、機械学習は教師あり学習と教師なし学習の 2 つの主なタイプに分類できます。

*教師あり学習*には、データの測定された特徴とデータに関連付けられたラベルとの関係を何らかの方法でモデル化することが含まれます。このモデルが決定されると、新しい未知のデータにラベルを適用するために使用できます。
これは、*分類* タスクと *回帰* タスクにさらに細分されます。分類では、ラベルは離散カテゴリですが、回帰では、ラベルは連続量です。
次のセクションでは、両方のタイプの教師あり学習の例を示します。

*教師なし学習*には、ラベルを参照せずにデータセットの特徴をモデル化することが含まれ、「データセット自体に語らせる」とよく言われます。
これらのモデルには、*クラスタリング*や*次元削減*などのタスクが含まれます。
クラスタリング アルゴリズムはデータの明確なグループを識別しますが、次元削減アルゴリズムはデータのより簡潔な表現を検索します。
次のセクションでは、教師なし学習の両方のタイプの例を見ていきます。

さらに、教師あり学習と教師なし学習の中間に位置する、いわゆる「半教師あり学習」メソッドがあります。
半教師あり学習法は、不完全なラベルしか利用できない場合に役立つことがよくあります。

## 機械学習アプリケーションの定性的な例

これらのアイデアをより具体的にするために、機械学習タスクの非常に単純な例をいくつか見てみましょう。
これらの例は、この章で検討する機械学習タスクのタイプの直感的で非定量的な概要を提供することを目的としています。
後のセクションでは、特定のモデルとその使用方法について詳しく説明します。
これらのより技術的な側面のプレビューについては、[Appendix: Figure Code](06.00-Figure-Code.ipynb) で次の図を生成する Python ソースを見つけることができます。


### 分類: 離散ラベルの予測

最初に、単純な*分類*タスクを見てみましょう。このタスクでは、一連のラベル付きポイントが与えられ、これらを使用してラベルのないポイントを分類したいと考えています。

次の図に示すデータがあるとします。

![](figures/05.01-classification-1.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Classification-Example-Figure-1)

ここには 2 次元データがあります。つまり、平面上の点の *(x,y)* 位置で表される、各点に 2 つの *特徴* があります。
さらに、各ポイントに 2 つの *クラス ラベル* のうちの 1 つがあり、ここではポイントの色で表されています。
これらの機能とラベルから、新しいポイントに「青」と「赤」のどちらのラベルを付けるかを決定できるモデルを作成したいと考えています。

このような分類タスクには多くの可能なモデルがありますが、ここでは非常に単純なものを使用します。 2 つのグループは、2 つのグループの間の平面を通る直線を引くことによって分離できると仮定し、線の両側の点が同じグループに収まるようにします。
ここで、*モデル* は「直線がクラスを分離する」というステートメントの定量的なバージョンであり、*モデル パラメーター* はデータのその線の位置と方向を表す特定の数値です。
これらのモデル パラメーターの最適値は、データから学習されます (これは、機械学習における「学習」です)。これは、多くの場合、*モデルのトレーニング* と呼ばれます。

次の図は、このデータのトレーニング済みモデルがどのように見えるかを視覚的に表現したものです。

![](figures/05.01-classification-2.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Classification-Example-Figure-2)

このモデルがトレーニングされたので、新しいラベルのないデータに一般化できます。
言い換えると、新しいデータ セットを取得し、このモデル ラインをそれを通して描画し、このモデルに基づいて新しいポイントにラベルを割り当てることができます。
この段階は通常*予測*と呼ばれます。次の図を参照してください。

![](figures/05.01-classification-3.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Classification-Example-Figure-3)

これは、機械学習における分類タスクの基本的な考え方であり、「分類」は、データに個別のクラス ラベルがあることを示します。
一見すると、これはかなり些細なことのように見えるかもしれません。このデータを見て、この分類を達成するためにそのような差別的な線を引くのは比較的簡単です。
ただし、機械学習アプローチの利点は、より多くの次元でより大きなデータセットに一般化できることです。

たとえば、これは電子メールの自動スパム検出のタスクに似ています。この場合、次の機能とラベルを使用できます。

- *機能 1*、*機能 2* など。重要な単語またはフレーズの $\to$ 正規化されたカウント (「バイアグラ」、「ナイジェリアの王子」など)
- *label* $\to$ "spam" または "not spam"

トレーニング セットの場合、これらのラベルは、少数の代表的な電子メール サンプルを個別に検査することによって決定される場合があります。残りの電子メールについては、モデルを使用してラベルが決定されます。
適切に構築された十分な機能 (通常、数千または数百万の単語またはフレーズ) を備えた適切にトレーニングされた分類アルゴリズムの場合、このタイプのアプローチは非常に効果的です。
このようなテキストベースの分類の例は、[In Depth: Naive Bayes Classification](05.05-Naive-Bayes.ipynb) で確認できます。

より詳細に説明するいくつかの重要な分類アルゴリズムは、ガウス ナイーブ ベイズ ([In Depth: Naive Bayes Classification](05.05-Naive-Bayes.ipynb) を参照)、サポート ベクター マシン ([In Depth: Naive Bayes Classification](05.05-Naive-Bayes.ipynb) を参照)、およびランダム フォレスト分類 ([In Depth: Naive Bayes Classification](05.05-Naive-Bayes.ipynb) を参照) です。

### 回帰: 連続ラベルの予測

分類アルゴリズムの離散ラベルとは対照的に、ラベルが連続量である単純な *回帰* タスクを次に見ていきます。

次の図に示すデータについて考えてみましょう。このデータは、それぞれに連続したラベルが付いた一連のポイントで構成されています。

![](figures/05.01-regression-1.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Regression-Example-Figure-1)

分類の例と同様に、2 次元データがあります。つまり、各データ ポイントを表す 2 つの特徴があります。
各ポイントの色は、そのポイントの連続ラベルを表します。

このタイプのデータに使用できる回帰モデルは多数ありますが、ここでは単純な線形回帰を使用してポイントを予測します。
この単純な線形回帰モデルは、ラベルを 3 番目の空間次元として扱う場合、平面をデータに適合させることができると想定しています。
これは、2 つの座標を持つデータに線を当てはめるというよく知られた問題をより高度に一般化したものです。

次の図に示すように、この設定を視覚化できます。

![](figures/05.01-regression-2.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Regression-Example-Figure-2)

ここでの *feature 1-feature 2* 平面は、前の 2 次元プロットと同じであることに注意してください。ただし、この場合は、ラベルを色と 3 次元の軸位置の両方で表しています。
この観点から、この 3 次元データに平面を当てはめることで、入力パラメーターの任意のセットに対して予想されるラベルを予測できるようになることは合理的であるように思われます。
2 次元投影に戻ると、このような平面を当てはめると、次の図に示す結果が得られます。

![](figures/05.01-regression-3.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Regression-Example-Figure-3)

この適合平面により、新しい点のラベルを予測するために必要なものが得られます。
視覚的には、次の図に示す結果が得られます。

![](figures/05.01-regression-4.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Regression-Example-Figure-4)

分類の例と同様に、次元数が少ない場合、これは些細なことに思えるかもしれません。
しかし、これらの方法の威力は、非常に多くの特徴を持つデータの場合に、簡単に適用して評価できることです。

たとえば、これは望遠鏡で観測された銀河までの距離を計算するタスクに似ています。この場合、次の機能とラベルを使用できます。

- *feature 1*, *feature 2*, など。いくつかの波長または色のいずれかでの各銀河の $\to$ の明るさ
- *label* 銀河の $\to$ 距離または赤方偏移

これらの少数の銀河の距離は、独立した一連の (通常はより費用のかかる) 観測によって決定される場合があります。
残りの銀河までの距離は、適切な回帰モデルを使用して推定できます。セット全体でより高価な観測を使用する必要はありません。
天文学界では、これは「測光赤方偏移」問題として知られています。

ここで説明する重要な回帰アルゴリズムには、線形回帰 ([In Depth: Linear Regression](05.06-Linear-Regression.ipynb) を参照)、サポート ベクター マシン ([In Depth: Linear Regression](05.06-Linear-Regression.ipynb) を参照)、およびランダム フォレスト回帰 ([In Depth: Linear Regression](05.06-Linear-Regression.ipynb) を参照) があります。

### クラスタリング: ラベルのないデータのラベルを推測する

先ほど見た分類と回帰の図は、新しいデータのラベルを予測するモデルを構築しようとしている教師あり学習アルゴリズムの例です。
教師なし学習には、既知のラベルを参照せずにデータを記述するモデルが含まれます。

教師なし学習の一般的なケースの 1 つは「クラスタリング」です。この場合、データはいくつかの個別のグループに自動的に割り当てられます。
たとえば、次の図に示すような 2 次元データがあるとします。

![](figures/05.01-clustering-1.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Clustering-Example-Figure-2)

肉眼で見ると、これらの点のそれぞれが異なるグループの一部であることは明らかです。
この入力が与えられると、クラスタリング モデルはデータの固有の構造を使用して、どのポイントが関連しているかを判断します。
非常に高速で直感的な *k*-means アルゴリズム ([In Depth: K-Means Clustering](05.11-K-Means.ipynb) を参照) を使用して、次の図に示すクラスターを見つけます。

![](figures/05.01-clustering-2.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Clustering-Example-Figure-2)

*k* - *k* 個のクラスター中心からなるモデルに適合することを意味します。最適な中心は、割り当てられた中心からの各ポイントの距離を最小化する中心であると想定されます。
繰り返しますが、これは 2 次元での些細な作業のように思えるかもしれませんが、データがより大きく複雑になるにつれて、このようなクラスタリング アルゴリズムを使用してデータセットから有用な情報を抽出できます。

*k*-means アルゴリズムについては、[In Depth: K-Means Clustering](05.11-K-Means.ipynb) で詳しく説明します。
その他の重要なクラスタリング アルゴリズムには、ガウス混合モデル ([In Depth: Gaussian Mixture Models](05.12-Gaussian-Mixtures.ipynb) を参照) やスペクトル クラスタリング ([In Depth: Gaussian Mixture Models](05.12-Gaussian-Mixtures.ipynb) を参照) などがあります。

### 次元削減: ラベルのないデータの構造を推測する

次元削減は、ラベルやその他の情報がデータセット自体の構造から推測される教師なしアルゴリズムの別の例です。
次元削減は、前に見た例よりも少し抽象的ですが、一般的に、完全なデータセットの関連する品質を何らかの方法で保持するデータの低次元表現を引き出すことを目指しています。
[In-Depth: Manifold Learning](05.10-Manifold-Learning.ipynb) で説明するように、さまざまな次元削減ルーチンがさまざまな方法でこれらの関連する品質を測定します。

この例として、次の図に示すデータを考えてみましょう。

![](figures/05.01-dimesionality-1.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Dimensionality-Reduction-Example-Figure-1)

視覚的には、このデータに何らかの構造があることは明らかです。これは、この 2 次元空間内でらせん状に配置された 1 次元の線から引き出されたものです。
この 1 次元データは高次元空間に埋め込まれていますが、ある意味では、このデータは "本質的に" 1 次元のみであると言えます。
この場合の適切な次元削減モデルは、この非線形埋め込み構造に敏感であり、この低次元表現を引き出すことができます。

次の図は、まさにこれを行う多様体学習アルゴリズムである Isomap アルゴリズムの結果の視覚化を示しています。

![](figures/05.01-dimesionality-2.png)
[figure source in Appendix](06.00-Figure-Code.ipynb#Dimensionality-Reduction-Example-Figure-2)

色 (抽出された 1 次元の潜在変数を表す) がスパイラルに沿って一様に変化することに注意してください。これは、アルゴリズムが実際に目で見た構造を検出したことを示しています。
前の例と同様に、次元削減アルゴリズムの能力は、高次元の場合により明確になります。
たとえば、100 個または 1,000 個の特徴を持つデータセット内の重要な関係を視覚化したい場合があります。
1,000 次元のデータを視覚化することは困難です。これをより扱いやすくする 1 つの方法は、次元削減技術を使用してデータを 2 次元または 3 次元に削減することです。

ここで説明するいくつかの重要な次元削減アルゴリズムは、主成分分析 ([In Depth: Principal Component Analysis](05.09-Principal-Component-Analysis.ipynb) を参照) と、Isomap や局所線形埋め込み ([In Depth: Principal Component Analysis](05.09-Principal-Component-Analysis.ipynb) を参照) を含むさまざまな多様体学習アルゴリズムです。

＃＃ まとめ

ここでは、機械学習アプローチのいくつかの基本的なタイプのいくつかの簡単な例を見てきました。
言うまでもなく、多くの重要な実用的な詳細について説明してきましたが、このセクションで、機械学習のアプローチで解決できる問題の種類についての基本的な考え方を理解していただければ幸いです。

要するに、次のことがわかりました。

- *教師あり学習*: ラベル付けされたトレーニング データに基づいてラベルを予測できるモデル

- *分類*: ラベルを 2 つ以上の個別のカテゴリとして予測するモデル
- *回帰*: 連続ラベルを予測するモデル

- *教師なし学習*: ラベル付けされていないデータの構造を識別するモデル

- *クラスタリング*: データ内の異なるグループを検出して識別するモデル
- *次元削減*: 高次元データ内の低次元構造を検出および識別するモデル

次のセクションでは、これらのカテゴリ内でさらに深く掘り下げ、これらの概念が役立つ場所の興味深い例をいくつか見ていきます。
