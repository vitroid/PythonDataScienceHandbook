# IPython: 通常の Python を超えて

Python の開発環境には多くのオプションがあり、自分の仕事でどれを使用しているのかよく尋ねられます。
私の答えは時々人々を驚かせます: 私の好みの環境は [IPython](http://ipython.org/) とテキスト エディター (私の場合は気分に応じて Emacs または Atom) です。
IPython (*Interactive Python* の略) は、2001 年に Fernando Perez によって強化された Python インタープリターとして開始され、Perez の言葉を借りれば、「研究用コンピューティングのライフサイクル全体のためのツール」を提供することを目的としたプロジェクトに成長しました。
Python がデータ サイエンス タスクのエンジンである場合、IPython は対話型のコントロール パネルと考えることができます。

IPython は、Python への便利な対話型インターフェイスであるだけでなく、言語に多くの便利な構文の追加も提供します。ここでは、これらの追加機能の中で最も役立つものについて説明します。
さらに、IPython は [Jupyter project](http://jupyter.org) と密接に結びついており、開発、コラボレーション、共有、さらにはデータ サイエンスの結果の公開に役立つブラウザー ベースのノートブックを提供します。
IPython ノートブックは、実際には、Julia、R、およびその他のプログラミング言語のノートブックを含む、より広範な Jupyter ノートブック構造の特殊なケースです。
ノートブック形式の有用性の例として、今読んでいるページを見てください。この本の原稿全体は、一連の IPython ノートブックとして構成されています。

IPython は、インタラクティブな科学的およびデータ集約型コンピューティングのために Python を効果的に使用することを目的としています。
この章ではまず、データ サイエンスの実践に役立つ IPython の機能のいくつかを順を追って説明し、特に Python の標準機能を超えて提供される構文に焦点を当てます。
次に、データ サイエンス コードの作成と使用における一般的なタスクを高速化できる、より便利な「マジック コマンド」のいくつかについて、もう少し詳しく説明します。
最後に、データの理解と結果の共有に役立つノートブックの機能のいくつかに触れます。

## シェルまたはノートブック?

この章で説明する IPython の主な使用方法は 2 つあります。IPython シェルと IPython ノートブックです。
この章の内容の大部分は両方に関連しており、最も便利なものに応じて例を切り替えます。
いずれかに関連するいくつかのセクションでは、その事実を明示的に述べます。
始める前に、IPython シェルと IPython ノートブックを起動する方法について説明します。

### IPython シェルの起動

この章は、本書の大部分と同様に、受動的に吸収されるようには設計されていません。
本書を読みながら、私たちが説明するツールと構文を試してみることをお勧めします。これを行うことで構築される筋肉の記憶は、それについて読むという単純な行為よりもはるかに役立ちます。
コマンドラインで **``ipython``** と入力して、IPython インタープリターを起動することから始めます。または、Anaconda や EPD などのディストリビューションをインストールしている場合は、システム固有のランチャーがある可能性があります (これについては、[Help and Documentation in IPython](01.01-Help-And-Documentation.ipynb) で詳しく説明します)。

これを行うと、次のようなプロンプトが表示されます。
```
IPython 4.0.1 -- An enhanced Interactive Python.
?         -> Introduction and overview of IPython's features.
%quickref -> Quick reference.
help      -> Python's own help system.
object?   -> Details about 'object', use 'object??' for extra details.
In [1]:
```
これで、フォローする準備が整いました。

### Jupyter ノートブックの起動

Jupyter ノートブックは、IPython シェルへのブラウザー ベースのグラフィカル インターフェイスであり、その上に豊富な動的表示機能を構築します。
Python/IPython ステートメントを実行するだけでなく、ノートブックを使用すると、ユーザーは書式設定されたテキスト、静的および動的な視覚化、数式、JavaScript ウィジェットなどを含めることができます。
さらに、これらのドキュメントは、他の人がドキュメントを開いて自分のシステムでコードを実行できるように保存できます。

IPython ノートブックは Web ブラウザー ウィンドウから表示および編集できますが、コードを実行するには、実行中の Python プロセスに接続する必要があります。
このプロセス (「カーネル」と呼ばれる) は、システム シェルで次のコマンドを実行することで開始できます。

```
$ jupyter notebook
```

このコマンドは、ブラウザーに表示されるローカル Web サーバーを起動します。
何をしているかを示すログをすぐに吐き出します。そのログは次のようになります。

```
$ jupyter notebook
[NotebookApp] Serving notebooks from local directory: /Users/jakevdp/PythonDataScienceHandbook
[NotebookApp] 0 active kernels 
[NotebookApp] The IPython Notebook is running at: http://localhost:8888/
[NotebookApp] Use Control-C to stop this server and shut down all kernels (twice to skip confirmation).
```

コマンドを発行すると、既定のブラウザーが自動的に開き、一覧表示されたローカル URL に移動します。
正確なアドレスはシステムによって異なります。
ブラウザが自動的に開かない場合は、ウィンドウを開いてこのアドレス (この例では *http://localhost:8888/*) を手動で開くことができます。
