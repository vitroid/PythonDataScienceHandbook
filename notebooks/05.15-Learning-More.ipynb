# その他の機械学習リソース

この章では、主に Scikit-Learn ライブラリ内のツールを使用して、Python での機械学習について簡単に説明しました。
この章は、多くの興味深い重要なアルゴリズム、アプローチ、および議論をカバーするには短すぎます。
ここで、機械学習に興味のある方のために、機械学習について詳しく学ぶためのリソースをいくつか提案したいと思います。

## Python での機械学習

Python での機械学習についてさらに学習するには、次のリソースのいくつかをお勧めします。

- [The Scikit-Learn website](http://scikit-learn.org): Scikit-Learn Web サイトには、ここで説明したモデルの一部をカバーする印象的なドキュメントと例が多数あります。最も重要で頻繁に使用される機械学習アルゴリズムの簡単な調査が必要な場合は、この Web サイトから始めることをお勧めします。

- *SciPy、PyCon、および PyData チュートリアル ビデオ*: Scikit-Learn およびその他の機械学習のトピックは、Python に焦点を当てた多くのカンファレンス シリーズ、特に PyCon、SciPy、および PyData カンファレンスのチュートリアル トラックで長年にわたって人気があります。簡単な Web 検索で最新のものを見つけることができます。

- [*Introduction to Machine Learning with Python*](http://shop.oreilly.com/product/0636920030515.do): Andreas C. Mueller と Sarah Guido によって書かれたこの本には、この章のトピックがより完全に扱われています。機械学習の基礎を確認し、Scikit-Learn ツールキットを限界まで押し上げることに興味がある場合、これは、Scikit-Learn チームで最も多くの成果を上げている開発者の 1 人によって書かれた優れたリソースです。

- [*Python Machine Learning*](https://www.packtpub.com/big-data-and-business-intelligence/python-machine-learning): Sebastian Raschka の本は、Scikit-learn 自体にはあまり焦点を当てておらず、Python で利用できる幅広い機械学習ツールに重点を置いています。特に、Python ベースの機械学習アプローチを大規模で複雑なデータセットに拡張する方法について、非常に有益な議論がいくつかあります。

## 一般的な機械学習

もちろん、機械学習は Python の世界だけではありません。知識をさらに深めるための優れたリソースが数多くあります。ここでは、役立つと思われるいくつかを紹介します。

- [*Machine Learning*](https://www.coursera.org/learn/machine-learning): Andrew Ng (Coursera) によって教えられた、これはアルゴリズムの観点から機械学習の基礎をカバーする非常に明確に教えられた無料のオンライン コースです。数学とプログラミングの学部レベルの理解を前提としており、最も重要な機械学習アルゴリズムのいくつかについて詳細に検討します。アルゴリズムによって採点される宿題では、実際にこれらのモデルのいくつかを自分で実装する必要があります。

- [*Pattern Recognition and Machine Learning*](http://www.springer.com/us/book/9780387310732): Christopher Bishop によって書かれたこの古典的な技術テキストは、この章で説明されている機械学習の概念を詳細にカバーしています。このテーマをさらに進める予定がある場合は、この本を本棚に置いておく必要があります。

- [*Machine Learning: a Probabilistic Perspective*](https://mitpress.mit.edu/books/machine-learning-0): Kevin Murphy によって書かれた、これは優れた大学院レベルのテキストであり、ほぼすべての重要な機械学習アルゴリズムをゼロから統合された確率論的観点から探究しています。

これらのリソースは、この本で紹介されている資料よりも技術的ですが、これらの方法の基礎を本当に理解するには、その背後にある数学を深く掘り下げる必要があります。
挑戦する準備ができていて、データ サイエンスを次のレベルに引き上げる準備ができている場合は、ためらわずに飛び込んでください。
