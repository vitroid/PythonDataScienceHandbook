## データサイエンスとは?

これは Python でデータ サイエンスを行うことについての本ですが、すぐに疑問が生じます: *データ サイエンス* とは何ですか?
特に、この用語がどこにでもあることを考えると、明確にするのは驚くほど難しい定義です.
口先だけの批評家は、この用語を不必要なラベル (結局のところ、どのような科学がデータを含まないのでしょうか?) または履歴書に塩を塗って熱狂的な技術採用担当者の目を引くためだけに存在する単純な流行語として、さまざまに却下しています。

私の考えでは、これらの批評は重要な何かを見逃しています。
データ サイエンスは、誇大広告で飾られているにもかかわらず、業界や学界全体の多くのアプリケーションでますます重要になっている学際的な一連のスキルに対して、おそらく私たちが持っている最高のラベルです。
この学際的な部分が重要です。私の考えでは、データ サイエンスの既存の最良の定義は、2010 年 9 月に彼のブログで最初に公開された Drew Conway のデータ サイエンス ベン図によって説明されています。

![Data Science Venn Diagram](figures/Data_Science_VD.png)

<small>(出典: [Drew Conway](http://drewconway.com/zia/2013/3/26/the-data-science-venn-diagram)。許可を得て使用。)</small>

交差点のラベルのいくつかは少し皮肉っぽいですが、この図は、人々が「データ サイエンス」と言うときに意味していると私が考えるものの本質を捉えています。それは基本的に*学際的な*主題です。
データ サイエンスは、重複する 3 つの異なる分野で構成されています。このデータを効率的に保存、処理、視覚化するためのアルゴリズムを設計および使用できる *コンピューター科学者* のスキル。 *ドメインの専門知識* - 主題の「古典的な」トレーニングと考えることができるもの - は、適切な質問を作成し、その回答を文脈に入れるために必要です。

このことを念頭に置いて、データ サイエンスを、学ぶべき新しい知識の領域ではなく、現在の専門分野に適用できる新しい一連のスキルと考えることをお勧めします。
選挙結果の報告、株式収益の予測、オンライン広告クリックの最適化、顕微鏡写真での微生物の特定、天体の新しいクラスの探索、または他の分野のデータの操作など、どのような目的であっても、この本の目標は、次のことを行う能力を提供することです。選択した主題分野について新しい質問をしたり、答えたりします。

## この本は誰のためのものですか?

ワシントン大学とさまざまなテクノロジに焦点を当てたカンファレンスやミートアップの両方で教えているときに、私が聞いた最も一般的な質問の 1 つは、「Python をどのように学習すればよいですか?」というものです。
質問者は一般的に技術志向の学生、開発者、または研究者であり、多くの場合、コードの記述や計算ツールや数値ツールの使用に関してすでに十分なバックグラウンドを持っています。
これらの人々のほとんどは、Python *それ自体*を学びたいとは思っていませんが、データ集約型の計算科学のツールとして使用する目的で言語を学びたいと考えています。
この視聴者向けのビデオ、ブログ投稿、チュートリアルの大規模なパッチワークがオンラインで入手できますが、私は長い間、この質問に対する適切な答えが 1 つもないことに不満を感じてきました。それがこの本にインスピレーションを与えたものです。

この本は、Python やプログラミング全般の紹介を意図したものではありません。読者は、関数の定義、変数の割り当て、オブジェクトのメソッドの呼び出し、プログラムの流れの制御、その他の基本的なタスクなど、Python 言語に精通していることを前提としています。
代わりに、Python ユーザーが Python のデータ サイエンス スタック (IPython、NumPy、Pandas、Matplotlib、Scikit-Learn、および関連ツールなどのライブラリ) を使用して、データを効果的に保存、操作、および洞察を得る方法を学習できるようにすることを目的としています。

## なぜ Python なのか?

Python は、大規模なデータセットの分析と視覚化を含む、科学計算タスクのための一流のツールとして、過去数十年にわたって登場しました。
これは、Python 言語の初期の支持者にとっては驚きだったかもしれません。言語自体は、データ分析や科学計算を念頭に置いて特別に設計されたわけではありません。
データ サイエンスにおける Python の有用性は、主にサードパーティ パッケージの大規模でアクティブなエコシステムに由来しています。コンピューティング タスク、出版品質のビジュアライゼーション用の *Matplotlib*、コードのインタラクティブな実行と共有用の *IPython*、機械学習用の *Scikit-Learn*、および次のページで説明するその他の多くのツール。

Python 言語自体のガイドを探している場合は、この本の姉妹プロジェクトである「[A Whirlwind Tour of the Python Language](https://github.com/jakevdp/WhirlwindTourOfPython)」をお勧めします。
この短いレポートでは、1 つまたは複数の他のプログラミング言語に既に精通しているデータ サイエンティストを対象として、Python 言語の重要な機能を紹介します。

### Python 2 対 Python 3

この本では、Python 3 の構文を使用しています。これには、Python の 2.x シリーズと互換性のない言語拡張機能が含まれています。
Python 3.0 は 2008 年に最初にリリースされましたが、特に科学および Web 開発コミュニティでは、採用は比較的遅かったです。
これは主に、重要なサードパーティのパッケージとツールキットの多くが新しい言語内部と互換性を持つようになるまでに時間がかかったことが原因です。
ただし、2014 年初頭以降、データ サイエンス エコシステムの最も重要なツールの安定版リリースは Python 2 と 3 の両方と完全に互換性があるため、この本では新しい Python 3 構文を使用します。
ただし、この本のコード スニペットの大部分は、変更を加えなくても Python 2 でも機能します。

## 本書の概要

この本の各章では、Python データ サイエンス ストーリーの基本的な部分に貢献する特定のパッケージまたはツールに焦点を当てています。

1. IPython および Jupyter: これらのパッケージは、多くの Python を使用するデータ サイエンティストが作業する計算環境を提供します。
2. NumPy: このライブラリは、Python で高密度のデータ配列を効率的に格納および操作するための ``ndarray`` を提供します。
3. Pandas: このライブラリは、Python でのラベル付き/列データの効率的な保存と操作のための ``DataFrame`` を提供します。
4. Matplotlib: このライブラリは、Python で柔軟な範囲のデータ視覚化機能を提供します。
5. Scikit-Learn: このライブラリは、最も重要で確立された機械学習アルゴリズムの効率的でクリーンな Python 実装を提供します。

PyData の世界は確かにこれら 5 つのパッケージよりもはるかに大きく、日々成長しています。
これを念頭に置いて、これらのページを通じて、Python でできることの限界を押し広げている他の興味深い取り組み、プロジェクト、およびパッケージへの参照を提供するために、あらゆる試みを行っています。
とはいえ、これら 5 つは現在、Python データ サイエンスの分野で行われている多くの作業の基本であり、エコシステムがそれらを中心に成長し続けても、これらは重要であり続けると予想しています。

## インストールに関する考慮事項

科学計算を可能にする Python と一連のライブラリのインストールは簡単です。このセクションでは、コンピュータをセットアップする際の考慮事項について概説します。

Python をインストールするにはさまざまな方法がありますが、データ サイエンスで使用するためにお勧めするのは、Windows、Linux、または Mac OS X のいずれを使用していても同様に機能する Anaconda ディストリビューションです。
Anaconda ディストリビューションには 2 つのフレーバーがあります。

- [Miniconda](http://conda.pydata.org/miniconda.html) は、Python インタープリター自体と、Python パッケージ向けのクロスプラットフォーム パッケージ マネージャーとして動作する ``conda`` と呼ばれるコマンドライン ツールを提供します。に精通。

- [Anaconda](https://www.continuum.io/downloads) には、Python と conda の両方が含まれており、さらに、科学計算向けの他のプリインストール パッケージのスイートがバンドルされています。このバンドルのサイズのため、インストールで数ギガバイトのディスク容量が消費されることが予想されます。

Anaconda に含まれるパッケージはどれも、Miniconda の上に手動でインストールすることもできます。このため、Miniconda から始めることをお勧めします。

開始するには、Miniconda パッケージをダウンロードしてインストールし (必ず Python 3 のバージョンを選択してください)、次に、この本で使用されているコア パッケージをインストールします。

```
[~]$ conda install numpy pandas scikit-learn matplotlib seaborn jupyter
```

テキスト全体を通して、Python の科学的エコシステムの他のより専門的なツールも利用します。通常、インストールは **``conda install packagename``** と入力するだけで簡単にできます。
conda 環境の作成と使用に関する情報を含む、conda の詳細については (これを強くお勧めします)、[conda's online documentation](http://conda.pydata.org/docs/) を参照してください。
