# その他のリソース

この章では、データ分析に Pandas を効果的に使用するための多くの基本について説明しました。
それでも、私たちの議論からは多くのことが省略されています。
Pandas の詳細については、次のリソースをお勧めします。

- [Pandas online documentation](http://pandas.pydata.org/): これは、パッケージの完全なドキュメントのソースです。ドキュメント内の例は生成された小さなデータセットになる傾向がありますが、オプションの説明は完全であり、さまざまな機能の使用法を理解するのに非常に役立ちます。

- [*Python for Data Analysis*](http://shop.oreilly.com/product/0636920023784.do) Wes McKinney (Pandas の最初の作成者) によって書かれたこの本には、Pandas パッケージについて、この章のスペースよりもはるかに多くの詳細が含まれています。特に、財務コンサルタントとしての彼のパンとバターであ​​った時系列のツールについて深く掘り下げます。この本には、現実世界のデータセットから洞察を得るために Pandas を適用する面白い例もたくさんあります。ただし、この本は数年前のものであり、Pandas パッケージには、この本で取り上げていない新機能が多数含まれていることに注意してください (ただし、2017 年の新版に注目してください)。

- [Stack Overflow](http://stackoverflow.com/questions/tagged/pandas): Pandas には非常に多くのユーザーがいるため、質問があればスタック オーバーフローで回答されている可能性があります。 Pandas の使用は、いくつかの Google-Fu があなたの親友である場合です。お気に入りの検索エンジンにアクセスして、遭遇した質問、問題、またはエラーを入力するだけです。スタック オーバーフロー ページで答えが見つかる可能性が高くなります。

- [Pandas on PyVideo](http://pyvideo.org/search?q=pandas): PyCon から SciPy、PyData に至るまで、多くのカンファレンスで Pandas 開発者やパワー ユーザーによるチュートリアルが取り上げられています。特に PyCon チュートリアルは、非常によく吟味されたプレゼンターによって提供される傾向があります。

これらのリソースと、この章で説明するウォークスルーを組み合わせて使用​​することで、Pandas を使用して、遭遇したデータ分析の問題に取り組む態勢が整うことを願っています!
