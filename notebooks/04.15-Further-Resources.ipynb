# その他のリソース

## Matplotlib リソース

本の 1 つの章で、Matplotlib で利用可能なすべての機能とプロット タイプをカバーすることは期待できません。
これまで見てきた他のパッケージと同様に、IPython のタブ補完機能とヘルプ機能 ([Help and Documentation in IPython](01.01-Help-And-Documentation.ipynb) を参照) を自由に使用すると、Matplotlib の API を調べるときに非常に役立ちます。
さらに、Matplotlib の [online documentation](http://matplotlib.org/) も参考になります。
特に、そのページにリンクされている [Matplotlib gallery](http://matplotlib.org/gallery.html) を参照してください。何百もの異なるプロット タイプのサムネイルが表示され、それぞれが生成に使用された Python コード スニペットを含むページにリンクされています。
このようにして、さまざまなプロット スタイルと視覚化手法を視覚的に調べて学習することができます。

Matplotlib の本の長さの扱いについては、Matplotlib のコア開発者である Ben Root によって書かれた [*Interactive Applications Using Matplotlib*](https://www.packtpub.com/application-development/interactive-applications-using-matplotlib) をお勧めします。

## その他の Python グラフィック ライブラリ

Matplotlib は最も有名な Python 視覚化ライブラリですが、他にも検討する価値のある最新のツールがあります。
ここでそれらのいくつかを簡単に説明します。

- [Bokeh](http://bokeh.pydata.org) は、非常に大きなデータセットやストリーミング データセットを処理できる高度にインタラクティブな視覚化を作成する Python フロントエンドを備えた JavaScript 視覚化ライブラリです。 Python フロントエンドは、Bokeh JS エンジンで解釈できる JSON データ構造を出力します。
- [Plotly](http://plot.ly) は、Plotly 社の同名のオープン ソース製品であり、Bokeh と精神が似ています。 Plotly はスタートアップの主力製品であるため、高いレベルの開発努力を受けています。図書館の利用は完全無料です。
- [Vispy](http://vispy.org/) は、非常に大規模なデータセットの動的な視覚化に重点を置いて積極的に開発されたプロジェクトです。 OpenGL をターゲットとし、コンピュータの効率的なグラフィック プロセッサを利用するように構築されているため、非常に大きく見事な視覚化をレンダリングできます。
- [Vega](https://vega.github.io/) および [Vega](https://vega.github.io/) は、宣言的なグラフィックス表現であり、データ視覚化の基本言語に関する長年の研究の成果です。リファレンス レンダリングの実装は JavaScript ですが、API は言語に依存しません。 [Vega](https://vega.github.io/) パッケージには開発中の Python API があります。 2016 年の夏の時点ではまだ完全には成熟していませんが、このプロジェクトが Python やその他の言語での視覚化に共通の基準点を提供する可能性に非常に興奮しています。
