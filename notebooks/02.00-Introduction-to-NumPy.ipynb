# NumPy の紹介

この章では、第 3 章とともに、Python でメモリ内データを効果的にロード、格納、および操作するための手法について概説します。
トピックは非常に広範です。データセットは、ドキュメントのコレクション、画像のコレクション、サウンド クリップのコレクション、数値測定のコレクションなど、さまざまなソースとさまざまな形式から取得できます。
この明らかな不均一性にもかかわらず、基本的にすべてのデータを数値の配列と考えることができます。

たとえば、画像 (特にデジタル画像) は、領域全体のピクセルの明るさを表す単純な数値の 2 次元配列と考えることができます。
サウンド クリップは、強度と時間の 1 次元配列と考えることができます。
テキストはさまざまな方法で数値表現に変換できます。たとえば、特定の単語または単語のペアの頻度を表す 2 進数などです。
データが何であれ、分析可能にするための最初のステップは、データを数値の配列に変換することです。
(このプロセスの具体的な例については、[Feature Engineering](05.04-Feature-Engineering.ipynb) で後述します)

このため、数値配列の効率的な格納と操作は、データ サイエンスを行うプロセスにとって絶対に不可欠です。
このような数値配列を処理するために Python が持っている特殊なツール、NumPy パッケージと Pandas パッケージ (第 3 章で説明) を見てみましょう。

この章では、NumPy について詳しく説明します。 NumPy (*Numerical Python* の略) は、高密度のデータ バッファーを格納および操作するための効率的なインターフェイスを提供します。
いくつかの点で、NumPy 配列は Python の組み込みの ``list`` 型に似ていますが、配列のサイズが大きくなるにつれて、NumPy 配列はより効率的なストレージとデータ操作を提供します。
NumPy 配列は、Python のデータ サイエンス ツールのほぼ全体のエコシステムの中核を形成しているため、データ サイエンスのどの側面に興味があるとしても、NumPy を効果的に使用するための学習に費やす時間は貴重です。

序文で概説したアドバイスに従い、Anaconda スタックをインストールした場合は、NumPy が既にインストールされており、準備が整っています。
自分でやるタイプの場合は、http://www.numpy.org/ にアクセスして、そこにあるインストール手順に従ってください。
完了したら、NumPy をインポートしてバージョンを再確認できます。
