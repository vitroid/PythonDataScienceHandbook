# その他の IPython リソース

この章では、IPython を使用してデータ サイエンス タスクを有効にする方法の表面をなぞっただけです。
印刷物と Web の両方で、さらに多くの情報を入手できます。ここでは、役立つと思われるその他のリソースをいくつか紹介します。

## Web リソース

- [The IPython website](http://ipython.org): IPython Web サイトは、ドキュメント、例、チュートリアル、およびその他のさまざまなリソースにリンクしています。
- [The nbviewer website](http://nbviewer.jupyter.org/): このサイトは、インターネットで入手可能な任意の IPython ノートブックの静的レンダリングを表示します。フロント ページには、他の人々が何のために IPython を使用しているかを参照できるいくつかのサンプル ノートブックが掲載されています。
- [A gallery of interesting Jupyter Notebooks](https://github.com/jupyter/jupyter/wiki/A-gallery-of-interesting-Jupyter-Notebooks/): この増え続けるノートブックのリストは、nbviewer を利用しており、IPython で実行できる数値解析の深さと幅を示しています。短い例やチュートリアルから本格的なコースやノートブック形式の書籍まで、すべてが含まれています。
- ビデオ チュートリアル: インターネットを検索すると、IPython に関する多くのビデオ録画チュートリアルが見つかります。特に、PyCon、SciPy、および PyData 会議で、IPython と Jupyter の主要な作成者であり保守者である Fernando Perez と Brian Granger によるチュートリアルを探すことをお勧めします。

## 本

- [*Python for Data Analysis*](http://shop.oreilly.com/product/0636920023784.do): Wes McKinney の本には、データ サイエンティストとしての IPython の使用をカバーする章が含まれています。資料の多くはここで説明した内容と重複していますが、別の視点が常に役立ちます。
- [*Learning IPython for Interactive Computing and Data Visualization*](https://www.packtpub.com/big-data-and-business-intelligence/learning-ipython-interactive-computing-and-data-visualization): Cyrille Rossant によるこの短い本は、データ分析に IPython を使用するための優れた入門書です。
- [*IPython Interactive Computing and Visualization Cookbook*](https://www.packtpub.com/big-data-and-business-intelligence/ipython-interactive-computing-and-visualization-cookbook): 同じく Cyrille Rossant によるこの本は、データ サイエンスに IPython を使用するためのより長く、より高度な取り扱いです。その名前にもかかわらず、それは IPython だけに関するものではなく、幅広いデータ サイエンスのトピックについてもある程度深く掘り下げています。

最後に、自分でヘルプを見つけることができることを思い出してください: IPython の ``?`` ベースのヘルプ機能 ([Help and Documentation in IPython](01.01-Help-And-Documentation.ipynb) で説明) は、よく使用し、頻繁に使用すると非常に便利です。
ここや他の場所で例を見ていくと、これを使用して、IPython が提供するすべてのツールに慣れることができます。
