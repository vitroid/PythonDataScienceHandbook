# 機械学習

多くの点で、機械学習は、データ サイエンスがより広い世界に現れるための主要な手段です。
機械学習は、データ サイエンスのこれらの計算スキルとアルゴリズム スキルがデータ サイエンスの統計的思考と出会う場所であり、その結果、効果的な理論ではなく効果的な計算である推論とデータ探索へのアプローチのコレクションが得られます。

「機械学習」という用語は、魔法の薬のように使われることがあります。*機械学習をデータに適用すれば、すべての問題が解決されます!*
ご想像のとおり、現実がこれほど単純であることはめったにありません。
これらの方法は信じられないほど強力ですが、効果的であるためには、各方法の長所と短所をしっかりと把握し、バイアスと分散、オーバーフィッティングとアンダーフィッティングなどの一般的な概念を把握してアプローチする必要があります。

この章では、主に Python の [Scikit-Learn](http://scikit-learn.org) パッケージを使用して、機械学習の実用的な側面に飛び込みます。
これは、機械学習の分野を包括的に紹介することを意図したものではありません。これは大きなテーマであり、ここで取り上げるよりも技術的なアプローチが必要です。
また、Scikit-Learn パッケージの使用に関する包括的なマニュアルを意図したものでもありません (これについては、[Further Machine Learning Resources](05.15-Learning-More.ipynb) にリストされているリソースを参照できます)。
むしろ、この章の目標は次のとおりです。

- 機械学習の基本的な語彙と概念を紹介します。
- Scikit-Learn API を紹介し、その使用例をいくつか示します。
- いくつかの最も重要な機械学習アプローチの詳細を深く掘り下げ、それらがどのように機能し、いつ、どこで適用できるかについて直感を養うこと。

この資料の多くは、私が PyCon、SciPy、PyData、およびその他のカンファレンスで何度か行った Scikit-Learn のチュートリアルとワークショップから引用されています。
次のページが明確になっているのは、ワークショップの参加者や共同講師の方々が、長年にわたってこの資料について貴重なフィードバックをくれたことによるものと思われます。

最後に、これらの主題のいずれかについてより包括的または技術的な扱いを求めている場合は、[Further Machine Learning Resources](05.15-Learning-More.ipynb) にいくつかのリソースと参考文献をリストしました。
