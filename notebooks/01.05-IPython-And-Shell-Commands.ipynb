# IPython とシェル コマンド

標準の Python インタープリターを使用して対話的に作業する場合、不満の 1 つは、複数のウィンドウを切り替えて Python ツールとシステム コマンド ライン ツールにアクセスする必要があることです。
IPython はこのギャップを埋め、シェル コマンドを IPython ターミナル内から直接実行するための構文を提供します。
魔法は感嘆符で起こります: 行の ``!`` の後に現れるものはすべて、Python カーネルではなく、システムのコマンドラインによって実行されます。

以下は、Linux や Mac OSX などの Unix ライクなシステムを使用していることを前提としています。
以下の例のいくつかは、デフォルトで異なるタイプのシェルを使用する Windows では失敗します (ただし、2016 年の Windows でのネイティブ Bash シェルの発表により、すぐにこれは問題ではなくなる可能性があります!)。
シェル コマンドに慣れていない場合は、常に優れた Software Carpentry Foundation によってまとめられた [Shell Tutorial](http://swcarpentry.github.io/shell-novice/) を確認することをお勧めします。

## シェルの簡単な紹介

シェル/ターミナル/コマンドラインの完全な紹介は、この章の範囲をはるかに超えていますが、初心者のためにここで簡単な紹介を提供します。
シェルは、コンピューターとテキストで対話する方法です。
1980 年代半ばに、Microsoft と Apple が現在ユビキタスなグラフィカル オペレーティング システムの最初のバージョンを導入して以来、ほとんどのコンピューター ユーザーは、慣れ親しんだメニューのクリックとドラッグ アンド ドロップ操作でオペレーティング システムを操作してきました。
しかし、オペレーティング システムは、これらのグラフィカル ユーザー インターフェイスよりずっと前に存在し、主に一連のテキスト入力によって制御されていました。プロンプトでユーザーがコマンドを入力すると、コンピューターはユーザーが指示したことを実行します。
これらの初期のプロンプト システムは、ほとんどのアクティブなデータ サイエンティストが現在も使用しているシェルとターミナルの前身です。

シェルに慣れていない人は、アイコンやメニューをクリックするだけで多くの結果が得られるのに、なぜこれを気にするのかと尋ねるかもしれません。
シェル ユーザーは別の質問に答えるかもしれません: 入力することでより簡単に物事を達成できるのに、なぜアイコンを探してメニューをクリックするのでしょうか?
典型的な技術的嗜好の行き詰まりのように聞こえるかもしれませんが、基本的なタスクを超えると、シェルが高度なタスクをはるかに制御できることがすぐに明らかになります。

例として、ユーザーがシステム上のディレクトリとファイルを探索、作成、および変更する Linux/OSX シェル セッションのサンプルを次に示します (``osx:~ $`` はプロンプトであり、``$ `` 記号は入力されたコマンドです; ``#`` が前にあるテキストは、実際に入力するものではなく、単なる説明を意味します):

```bash
osx:~ $ echo "hello world"             # echo is like Python's print function
hello world

osx:~ $ pwd                            # pwd = print working directory
/home/jake                             # this is the "path" that we're sitting in

osx:~ $ ls                             # ls = list working directory contents
notebooks  projects 

osx:~ $ cd projects/                   # cd = change directory

osx:projects $ pwd
/home/jake/projects

osx:projects $ ls
datasci_book   mpld3   myproject.txt

osx:projects $ mkdir myproject          # mkdir = make new directory

osx:projects $ cd myproject/

osx:myproject $ mv ../myproject.txt ./  # mv = move file. Here we're moving the
                                        # file myproject.txt from one directory
                                        # up (../) to the current directory (./)
osx:myproject $ ls
myproject.txt
```

これらはすべて、アイコンやメニューをクリックするのではなくコマンドを入力して、使い慣れた操作 (ディレクトリ構造のナビゲート、ディレクトリの作成、ファイルの移動など) を実行するコンパクトな方法であることに注意してください。
いくつかのコマンド (``pwd``、``ls``、``cd``、``mkdir``、および ``cp``) だけで、最も一般的なファイル操作の多くを実行できることに注意してください。
シェル アプローチが本当に強力になるのは、これらの基本を超えたときです。

## IPython のシェル コマンド

コマンドラインで機能するすべてのコマンドは、先頭に ``!`` 文字を付けることで IPython で使用できます。
たとえば、 ls 、 pwd 、および echo コマンドは次のように実行できます。

```ipython
In [1]: !ls
myproject.txt

In [2]: !pwd
/home/jake/projects/myproject

In [3]: !echo "printing from the shell"
printing from the shell
```

## シェルとの間で値を渡す

シェル コマンドは、IPython から呼び出すことができるだけでなく、IPython 名前空間とやり取りするようにすることもできます。
たとえば、代入演算子を使用して、任意のシェル コマンドの出力を Python リストに保存できます。

```ipython
In [4]: contents = !ls

In [5]: print(contents)
['myproject.txt']

In [6]: directory = !pwd

In [7]: print(directory)
['/Users/jakevdp/notebooks/tmp/myproject']
```

これらの結果はリストとして返されるのではなく、IPython で定義された特別なシェルの戻り値の型として返されることに注意してください。

```ipython
In [8]: type(directory)
IPython.utils.text.SList
```

これは、Python のリストとよく似た外観と機能を備えていますが、次のような追加機能があります。
``grep`` および ``fields`` メソッドと、便利な方法で結果を検索、フィルタリング、および表示できる ``s``、``n``、および ``p`` プロパティ。
これらの詳細については、IPython の組み込みのヘルプ機能を使用できます。

``{varname}`` 構文を使用して、Python 変数をシェルに渡すという逆方向の通信が可能です。

```ipython
In [9]: message = "hello from Python"

In [10]: !echo {message}
hello from Python
```

中括弧には変数名が含まれており、シェル コマンドで変数の内容に置き換えられます。

# シェル関連の魔法コマンド

IPython のシェル コマンドをしばらく試してみると、!cd を使用してファイル システムをナビゲートできないことに気付くかもしれません。

```ipython
In [11]: !pwd
/home/jake/projects/myproject

In [12]: !cd ..

In [13]: !pwd
/home/jake/projects/myproject
```

その理由は、ノートブックのシェル コマンドが一時的なサブシェルで実行されるためです。
作業ディレクトリをより永続的な方法で変更したい場合は、 %cd 魔法のコマンドを使用できます:

```ipython
In [14]: %cd ..
/home/jake/projects
```

実際、デフォルトでは ``%`` 記号なしでこれを使用することさえできます:

```ipython
In [15]: cd myproject
/home/jake/projects/myproject
```

これは ``automagic`` 関数として知られており、この動作は ``%automagic`` マジック関数で切り替えることができます。

``%cd`` の他に利用可能なシェルのようなマジック関数は ``%cat``, ``%cp``, ``%env``, ``%ls``, ``%man`` です, ``%mkdir``, ``%more``, ``%mv``, ``%pwd``, ``%rm``, ``%rmdir``, どれでも使用できます``automagic`` がオンの場合、``%`` 記号なし。
これにより、IPython プロンプトを通常のシェルのように扱うことができます。

```ipython
In [16]: mkdir tmp

In [17]: ls
myproject.txt  tmp/

In [18]: cp myproject.txt tmp/

In [19]: ls tmp
myproject.txt

In [20]: rm -r tmp
```

Python セッションと同じターミナル ウィンドウ内からシェルにアクセスできるということは、Python コードを記述するときにインタープリターとシェルの間を行き来する回数が大幅に減ることを意味します。
